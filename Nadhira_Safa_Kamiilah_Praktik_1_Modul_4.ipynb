## Mengatasi Missing Value

```python
#mengatasi missing value pada kolom name
#keputusan: menghapus data missing value pada kolom name karena jumlahnya relatif sedikit
df = df.dropna(subset=['name'])
```

```python
#mengatasi missing value pada kolom host_name
#keputusan: menghapus data missing value pada kolom host_name karena jumlahnya relatif sedikit
df = df.dropna(subset=['host_name'])
```

```python
#mengatasi missing value pada kolom price
#keputusan: menghapus data missing value pada kolom price karena jumlahnya relatif sedikit
df = df.dropna(subset=['price'])
```

```python
#mengatasi missing value pada kolom number_of_reviews
#keputusan: mengisi missing value dengan nilai median karena nilai tengah membuat persebarannya lebih merata
#median juga robust terhadap outlier
#alasan imputasi: kolom number_of_review memiliki missing value cukup besar, ketika dihapus maka kemungkinan hasil analisis akan bias

#menghitung nilai median
median_nor = df['number_of_reviews'].median()

#mengisi missing values dengan nilai median
df['number_of_reviews'].fillna(median_nor, inplace=True)
```

```python
#mengatasi missing value pada kolom last_review
#mengisi missing value dengan modus
#alasan imputasi: kolom last_review memiliki missing value cukup besar, ketika dihapus maka kemungkinan hasil analisis akan bias

#mencari nilai modus
mode_nor = df['number_of_reviews'].mode()

#mengisi missing values dengan nilai modus
df['last_review'].fillna(mode_nor, inplace=True)
```

```python
#mengatasi missing value pada kolom reviews_per_month
#keputusan: mengisi missing value dengan nilai median karena nilai tengah membuat persebarannya lebih merata
#median juga robust terhadap outlier
#alasan imputasi: kolom reviews_per_month memiliki missing value cukup besar, ketika dihapus maka kemungkinan hasil analisis akan bias

#menghitung nilai median
median_rpm = df['reviews_per_month'].median()

#mengisi missing values dengan nilai median
df['reviews_per_month'].fillna(median_rpm, inplace=True)
```

## Mengatasi Outliers Data

```python
# membuat fungsi cek outlier. Lengkapi bagian ... !
def cek_outlier(nama_kolom, batas):
  Q1 = np.percentile(df[nama_kolom], 25,
                     interpolation = 'midpoint')

  Q3 = np.percentile(df[nama_kolom], 75,
                     interpolation = 'midpoint')

  IQR = Q3-Q1

  if batas == 'upper_bound':
      bound = np.where(df[nama_kolom] >= (Q3+1.5*IQR))
  elif batas == 'lower_bound':
      bound = np.where(df[nama_kolom] <= (Q1-1.5*IQR))
  return list(bound[0])
```

> Pada hal ini dilakukan pengecekan outlier pada kolom `price`. Alasan memilih kolom price karena variabel ini akan memiliki peran penting dalam analisis lebih lanjut pada suatu perusahaan properti

```python
# Membuat boxplot dari kolom price
# Terlihat banyak titik observasi di luar Q3 yang menandakan banyak data yang terindikasi pencilan
plt.figure(figsize = (5,7))
sns.boxplot(y='price',data = df)
plt.show()
```

```python
# Cek outlier kolom price
upper_price = cek_outlier('price','upper_bound')
lower_price= cek_outlier('price','lower_bound')
print(f"Jumlah total outlier di kolom price : {len(upper_price) + len(lower_price)}")
```

# Bagian 3 : Analisa Data

```python
# Menampilkan harga rata-rata, termurah dan termahal
# yang disediakan per jenis tipe kamar beserta karakteristiknya
# Mengelompokkan data berdasarkan 'Room_type' dan menghitung statistik
# Membuat pivot table
pivot_table = pd.pivot_table(df, index='room_type', values=['price', 'minimum_nights', 'number_of_reviews'],
                             aggfunc={'price': ['mean', 'max', 'min'],
                                      'minimum_nights': 'mean',
                                      'number_of_reviews': 'mean'})

# Mengganti nama kolom agar sesuai dengan yang diminta
pivot_table.columns = ['rata-rata minimum pemesanan (malam)', 'rata-rata total review','harga max','harga rata-rata', 'harga min']

# Menampilkan hasil
print(pivot_table)
```

```python
# Mencari Total properti dan rata-rata harga per neighbourhood group
# Membuat pivot table
pivot_table = pd.pivot_table(df, index='neighbourhood_group', values='price',
                             aggfunc={'price': ['count', 'mean']})

# Mengganti nama kolom agar sesuai dengan yang diminta
pivot_table.columns = ['total_properti', 'avg_price_properti']

# Menampilkan hasil
print(pivot_table)
```

```python
# Dari tahun 2017-2019, berapa rata-rata reviews per month dan
# rata-rata availability kamar nya untuk setiap jenis tipe kamar per tahunnya
# Mengekstrak tahun dari kolom "last_review"
df['year'] = pd.to_datetime(df['last_review']).dt.year

# Membuat pivot table
pivot_table = pd.pivot_table(df, index=['year', 'room_type'],
                             values=['reviews_per_month', 'availability_365'],
                             aggfunc={'reviews_per_month': 'mean', 'availability_365': 'mean'})

# Mengganti nama kolom agar sesuai dengan yang diminta
pivot_table.columns = ['avg_review_per_month', 'avg_availability']

# Menampilkan hasil
print(pivot_table)
```
